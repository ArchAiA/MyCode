```python
#To figure out where we reach the 50% mark
#We can use numpy's searchsorted
#Note .values is needed in the newer versions of pandas so we don't 
#   return an array
df.prop.values.cumsum().searchsorted(0.5)

#So in 2008, 127 names made up 50% of the name base
```

```python
#Lets make a function to do the above process for a give data frame
#Feed in a data frame, pick the percent we want to look at (default 0.5)
def get_quantile_count(group, quantile = 0.5):
    
    #Make a df that sorts the input df by prop
    df = group.sort_values(by='prop', ascending=False)
    
    #Return the point which cumsum of prop passes the quantile specified 
    return df.prop.values.cumsum().searchsorted(quantile)

boys.groupby('year').apply(get_quantile_count).plot()
```

```python
import matplotlib.pyplot as plt

def get_quantile_count(group, quantile=0.5): 
    #Could group by sound examples first to ignore different spellings
    group = group.groupby('soundex').sum()
    df = group.sort_values(by='prop', ascending=False)
    return df.prop.values.cumsum().searchsorted(quantile)

#Lets look at 25% quantile for boys and girls separately
q = 0.25

#To pass additional function inpust with apply, add arguements to apply
#Alternatively coud use a lamda like
# f = lambda x: get_quantile_count(x,q)
#Or a partial func object

boy_ct = boys.groupby('year').apply(get_quantile_count, quantile=q) 
girl_ct = girls.groupby('year').apply(get_quantile_count, quantile=q)
boy_ct.plot(label='boy')
girl_ct.plot(label='girl')
plt.legend(loc='best')
```

<h1> Lets look at new csv file about the number of births per year that we want to join with the old one</h1>

```python
births = read_csv('births.csv')
```

```python
#Pandas version of join is "merge"
#Can specify left, right, inner/outer, sort or not, etc.. look at the doc

#Here, year and sex are included in both csv, so join on that
merged = merge(names, births, on=['year', 'sex'])  
```

```python
# same as above. Join is performed using the common columns: year and sex
# Many to many joins in SQL compute the cartesians products of duplicated keys.

#This is the same as above because it uses the intersection of the columns
#if on isn't specified (it does a cross or full inner join)
merge(names, births)  
```

```python
#Number of people with each name
merged['persons'] = np.floor(merged.prop * merged.births)
```

```python
merged.head()
```

```python
#Group by name and sex, and sum over persons -- gives total number of people
#with that name over all year
merged.groupby(['name', 'sex'])['persons'].sum() 
```

```python
#Same as above, but order by the result
merged.groupby(['name', 'sex'])['persons'].sum().sort_values()
```

```python
#Merge with boys only instead
mboys = merge(boys, births)  # inner join by default.
```

```python
mboys['persons'] = np.floor(mboys.prop * mboys.births)
```

```python
persons = mboys.set_index(['year', 'name']).persons   # Select out persons
```

```python
type(persons)  # One dimensional series.
```
