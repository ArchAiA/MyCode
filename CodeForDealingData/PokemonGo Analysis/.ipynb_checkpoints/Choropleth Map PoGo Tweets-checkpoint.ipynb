<h2> Converting JSON data to GeoJSON format </h2>

```python
import json
import pandas as pd
import bokeh
from bokeh.plotting import figure, output_notebook, show

#Inline Plotting
output_notebook()
 
```

```python
fname = "PoGo_Historical2.json"

# Tweets are stored in "fname" json file from live stream above
# We use the coordinates JSON information, since geo is depricated
with open(fname, 'r') as f:
    geo_data = {
        "type": "FeatureCollection",
        "features": []
    }
    for line in f:
        tweet = json.loads(line)
        if 'coordinates' in tweet:
            if tweet['coordinates']:
                geo_json_feature = {
                    "type": "Feature",
                    "geometry": tweet['coordinates'],
                    "properties": {
                        "text": tweet['text'],
                        "created_at": tweet['created_at']
                    }
                }
                geo_data['features'].append(geo_json_feature)
 
# Save geo data
with open('PoGo_geo_data.json', 'w') as fout:
    fout.write(json.dumps(geo_data, indent=4))
    
# SHOULD DELETE THE ABOVE FILE FIRST!!!
```

```python
def load_twitter_data(tweets_data_path):
    tweets_data = []
    tweets_file = open(tweets_data_path, "r")
    for line in tweets_file:
        try:
            tweet = json.loads(line)
            tweets_data.append(tweet)
        except:
            continue
    return tweets_data

def pop_tweets(path):
    tweets_data = load_twitter_data(path)
    tweets = pd.DataFrame(columns=['text','latt','long','location'])

    for tweet in tweets_data:
        if ('text' in tweet): 
            if tweet['coordinates'] != None:
                tweets.loc[len(tweets)]=[tweet['text'],tweet['coordinates']['coordinates'][0], \
                                         tweet['coordinates']['coordinates'][1],tweet['place']['name']]    
            else:
                tweets.loc[len(tweets)]=[tweet['text'],float('nan'),float('nan'),tweet['place']['name']]    
        
    return tweets
```

```python
PoGo_tweets = pop_tweets('PoGo_Historical2.json')
```

<h1> Converting from RYB to RGB </h1>
Following: http://vis.computer.org/vis2004/DVD/infovis/papers/gossett.pdf <br \>
http://math.stackexchange.com/questions/305395/ryb-and-rgb-color-space-conversion <br \>
Note that figure 3 has a typo - purple point should be (0.5,0,0.5)    
    

```python
def RYB_to_RGB(r,y,b): 
    '''
    #Original
    f000=np.array([1,1,1])
    f001=np.array([0.163,0.373,0.6])
    f010=np.array([1,1,0])
    f011=np.array([0,0.66,0.2])
    f100=np.array([1,0,0])
    f101=np.array([0.5,0,0.5])
    f110=np.array([1,0.5,0])
    f111=np.array([0.2,0.094,0.0])
    '''
    
    #Modified
    f000=np.array([0,0,0])
    f001=np.array([0,0,1])
    f010=np.array([1,1,0])
    f011=np.array([0,1,0])
    f100=np.array([1,0,0])
    f101=np.array([1,0,1])
    f110=np.array([1,0.647,0])
    f111=np.array([1,1,1])
    
    rgb=f000*(1-r)*(1-y)*(1-b) +\
        f001*(1-r)*(1-y)*b +\
        f010*(1-r)*y*(1-b) +\
        f100*r*(1-y)*(1-b) +\
        f011*(1-r)*y*b + \
        f101*r*(1-y)*b + \
        f110*r*y*(1-b) + \
        f111*r*y*b
    
    return rgb
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Loop through the x,y data, and count number of blue, yellow, and red team for each bin

#Set up binning of the map
num_x_bins=50
num_y_bins=50
x_step=(max(PoGo_tweets.x_coord)-min(PoGo_tweets.x_coord))/num_x_bins
y_step=(max(PoGo_tweets.y_coord)-min(PoGo_tweets.y_coord))/num_y_bins
x_bins=np.arange(min(PoGo_tweets.x_coord)+x_step,max(PoGo_tweets.x_coord),x_step) #marks the max of the bin
y_bins=np.arange(min(PoGo_tweets.y_coord)+y_step,max(PoGo_tweets.y_coord),y_step)

#Preallocate the fractional map
frac_map=np.zeros((num_y_bins,num_x_bins,4)) #Num xbins, Num Ybins, pixel RGBA

#Team color cuts

tweet_texts = PoGo_tweets['text'].str.lower()

red_cut    = [('#teamred' in DF_entry) or ('#teamvalor' in DF_entry) or \
                  ('#valor' in DF_entry) for DF_entry in tweet_texts]

blue_cut   = [('#teamblue' in DF_entry) or ('#teammystic' in DF_entry) or \
                  ('#mystic' in DF_entry) for DF_entry in tweet_texts]  

yellow_cut = [('#teamyellow' in DF_entry) or ('#teaminstinct' in DF_entry) or \
                  ('#instinct' in DF_entry) for DF_entry in tweet_texts]

total_iden = len(PoGo_tweets.ix[(pd.Series(red_cut) | pd.Series(blue_cut) | pd.Series(yellow_cut))])

#Loop through each bin of the map
current_xbin=0
for x_max in x_bins:
    current_ybin=0
    for y_max in y_bins:
        
        #XY cuts
        xy_cut   = (PoGo_tweets['x_coord'] > x_max-x_step) & (PoGo_tweets['x_coord'] <= x_max) & \
                   (PoGo_tweets['y_coord'] > y_max-y_step) & (PoGo_tweets['y_coord'] <= y_max)
                 
        #Fration of tweers in this bin that mention each team
        total_iden_bin  = len(PoGo_tweets.ix[xy_cut & (pd.Series(red_cut) | pd.Series(blue_cut) | pd.Series(yellow_cut))])
        
        #Default to white if there is no data
        if total_iden_bin > 0:
            frac_red    = len(PoGo_tweets.ix[xy_cut & red_cut])/total_iden_bin
            frac_yellow = len(PoGo_tweets.ix[xy_cut & yellow_cut])/total_iden_bin
            frac_blue   = len(PoGo_tweets.ix[xy_cut & blue_cut])/total_iden_bin
            ppl_per_bin = (13.1/(num_x_bins*num_y_bins/2500)) * (total_iden/512)   #new_num_people/new_num_bins = (old_num_people/old_num_bins)*(old_num_bins/new_num_bins)*(new_num_people/old_num_people)
            alpha_scale = 1 #total_iden_bin/ppl_per_bin #total_iden_bin/2 #just picking expectation of 10 counts for now.. need to sort this out
        else: 
            frac_red    = 1 #Set to zero if using the original RYB->RGB Conversion Cube
            frac_yellow = 1
            frac_blue   = 1
            alpha_scale = 1
        
        #Converting yellow to blue + green
        rgb=RYB_to_RGB(frac_red,frac_yellow,frac_blue)
        
        #Filling in the frac_map
        frac_map[current_ybin,current_xbin,:] = rgb[0],rgb[1],rgb[2],alpha_scale
        
        current_ybin += 1

    #Move onto next xbin        
    current_xbin += 1
        
```

<h1> Backup method for converting RYB -> RGB </h1>

```python
# elementary_colors.py
from math import degrees, radians, atan2, sin, cos

red = (255, 0, 0)
orange = (255, 160, 0)
yellow = (255, 255, 0)
green = (0, 255, 0)
cyan = (0, 255, 255)
blue = (0, 0, 255)
magenta = (255, 0, 255)
white = (255, 255, 255)
black = (0, 0, 0)
mud = (94, 81, 74)

colorwheel = [(0, red), (60, orange), (120, yellow), (180, green),
              (215, cyan), (250, blue), (330, magenta), (360, red)]

red_x, red_y = cos(radians(0)), sin(radians(0))
yellow_x, yellow_y = cos(radians(120)), sin(radians(120))
blue_x, blue_y = cos(radians(240)), sin(radians(240))

def lerp(left, right, left_part, total):
    if total == 0:
        return left
    ratio = float(left_part) / total
    return [l * ratio + r * (1.0 - ratio) for l,r in zip(left, right)]

def hue_to_rgb(deg):
    deg = deg % 360
    previous_angle, previous_color = colorwheel[0]
    for angle, color in colorwheel:
        if deg <= angle:
            return lerp(previous_color, color, angle - deg, angle - previous_angle)
        previous_angle = angle
        previous_color = color

def int_rgb(rgb):
    return tuple(int(c * 255.99 / 255) for c in rgb)

def rybwk_to_rgb(r, y, b, w, k):
    if r == 0 and y == 0 and b == 0:
        rgb = white
    else:
        hue = degrees(atan2(r * red_y + y * yellow_y + b * blue_y,
                            r * red_x + y * yellow_x + b * blue_x))
        rgb = hue_to_rgb(hue)
        rgb = lerp(mud, rgb, min(r, y, b), max(r, y, b))
    gray = lerp(white, black, w, w+k)
    rgb = lerp(rgb, gray, r+y+b, r+y+b+w+k)
    return int_rgb(rgb)
```
